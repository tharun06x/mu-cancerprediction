```python
data.loc[(data['age']<=0) | (data['age']>120),'age']=np.nan
data.loc[(data['bmi']<10)|(data['bmi']>60),'bmi']=np.nan
data.loc[(data['hypertension']<0)|(data['hypertension']>1),'hypertension']=np.nan
data.loc[(data['asthma']<0)|(data['asthma']>1),'asthma']=np.nan
data.loc[(data['other_cancer']<0)|(data['other_cancer']>1),'other_cancer']=np.nan
data.loc[(data['survived']<0)|(data['survived']>1),'survived']=np.nan
```

```python
skewed=['hypertension','cirrhosis','other_cancer','asthma']
sym=['age','bmi','cholesterol_level']
binary_col=[['Male','Female'],['No','Yes'],['Stage I','Stage II','Stage III','Stage IV'],
 ['Never Smoked','Passive Smoker','Former Smoker','Current Smoker']]
ord_col=['gender','family_history','cancer_stage','smoking_status']
```

```python
start=pd.to_datetime(data['diagnosis_date'])
end=pd.to_datetime(data['end_treatment_date'])
data['total_treatment_days']=(end-start).dt.days
data.drop(['diagnosis_date','end_treatment_date'],axis=1,inplace=True)
data['survived']=pd.to_numeric(data['survived'],errors='coerce')
data=data.dropna(subset=['survived'])
data['survived'] = data['survived'].astype(int)
```

```python

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
ordinalcol=Pipeline([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('encode',OrdinalEncoder(categories=binary_col))
])
onehotcol=Pipeline([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('encode',OneHotEncoder(drop='first',sparse_output=False))
])
preprocessing=ColumnTransformer([
    ('skewedcol',SimpleImputer(strategy="most_frequent"),skewed),
    ('normalcol',SimpleImputer(strategy='mean'),sym),
    ('ordinalenc',ordinalcol,ord_col),
    ('onehotenc',onehotcol,['treatment_type'])
],remainder='passthrough')

```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X=data.drop(['survived'],axis=1)
y=data['survived']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=42,stratify=y)
full_pipeline=Pipeline([
    ('prepro',preprocessing),
    ('addingsamples',SMOTE(random_state=42)),
    ('model',RandomForestClassifier(random_state=42,max_depth=10, n_estimators=1000))
])
full_pipeline.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,roc_auc_score
y_pred=full_pipeline.predict(X_test)
y_pred_proba=full_pipeline.predict_proba(X_test)[:,1]
print(classification_report(y_test,y_pred))
print(f"ROC_AUC SCORE:{roc_auc_score(y_test,y_pred_proba)}")
```
